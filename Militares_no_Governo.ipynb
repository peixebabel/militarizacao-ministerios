```python
import pandas as pd
import numpy as np

df = pd.read_csv('frac_militares_por_orgao_e_mes_ateh_2020-09.csv')
# df['orgao'].loc[2] = "Ciência & Comunic."
df
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px

def fn_showgrid(axis, bool_):
  axis.showgrid = bool_
  return axis

fig = make_subplots(rows=4, cols=4, shared_yaxes=False,
                    subplot_titles=(df['orgao']))
colors = px.colors.qualitative.Plotly + px.colors.qualitative.G10
colors.pop(7)
colors.pop(8)
# np.random.shuffle(colors)

for i in range(len(df)):
  sub_i = (i // 4) + 1
  sub_j = (i % 4) + 1

  y = df.iloc[i, 2:]

  df_mini = pd.DataFrame(df.iloc[i, 2:].T)
  df_mini['data'] = df_mini.index
  df_mini.columns = ['Percentual', 'Data']
  df_mini

  pfig = px.area(df_mini, x='Data', y='Percentual', color_discrete_sequence =[colors[i]])
  fig.add_trace(pfig.data[0], row=sub_i, col=sub_j)


fig.update_layout(height=700, template='plotly_white', xaxis=dict(showgrid=False),
                  title_text="Percentual de Militares comissionados no Governo" ) 
fig.for_each_xaxis(lambda axis: axis.tickfont.update(dict(size=8)))
fig.for_each_xaxis(lambda axis: fn_showgrid(axis, False) ) 
fig.for_each_annotation(lambda trace: trace.update(font=dict(size=14))) 
fig.add_vrect(x0="01/2019", x1="09/2020", line_width=0, fillcolor="black", opacity=0.2, layer="below",
              annotation_text="Bolsonaro", annotation_position="top left", annotation=dict(font_size=12))
fig.add_vrect(x0="05/2016", x1="12/2018", line_width=0, fillcolor="black", opacity=0.1, layer="below",
              annotation_text="Temer", annotation_position="top left", annotation=dict(font_size=12))
fig.update_yaxes(range=[0, 19])

fig.show()
```
